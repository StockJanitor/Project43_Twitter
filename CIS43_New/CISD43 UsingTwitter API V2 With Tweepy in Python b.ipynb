**Search use case**


```python
#import tweepy
#import the file that has your tokens and access keys
import tweepy
import Credentials
```

```python
#Initialize a client. The client is the object that has the different methods that you can use to connect to this Twitter API
#and pass it the bearer token

client=tweepy.Client(bearer_token=Credentials.bearer_token)

```

```python
# Search Recent Tweets

# This endpoint/method returns Tweets from the last seven days

response = client.search_recent_tweets("Tweepy")
# The method returns a Response object, a named tuple with data, includes,
# errors, and meta fields
print(response.meta)

# In this case, the data field of the Response returned is a list of Tweet
# objects
tweets = response.data

# Each Tweet object has default ID and text fields
for tweet in tweets:
    print(tweet.id)
    print(tweet.text)

# By default, this endpoint/method returns 10 results
# You can retrieve up to 100 Tweets by specifying max_results
response = client.search_recent_tweets("Tweepy", max_results=30)
```

```python
#Searching
#search_recent_tweets will require a query. query is where you specify what you are looking for.
#If you want to search tweets by keyword
#let us say you want the tweets about covid-- query="covid"
#you can have or condition-- query ="covid OR covid19"
#you can specify and condition by using a space between the terms-- query="covid covid19"
#tweets that contain both of covid and covid19
#you can add a parameter is:retweet to get tweets that are retweets
#useful link for queries: 

#let us get tweets about covid-- i.e using the keyword covid and that are not retweets
#mean that they original not retweets
#by default you get tweets back if you want more you can add a parameter max_results=30
query="covid -is:retweet"
response=client.search_recent_tweets(query=query, max_results=30)

print(response)
#the response has different things it has data, includes, errors, and meta.  bata y default it gives in the tweetid and text
#if I want something additional, I have to go ahead and do something extra
```

```python
#let us get only the tweets ids
for tweet in response.data:
    print(tweet.id)
```

```python
#As we saw you get the id and text
#let us say we want to get what language is this tweet was in.
# you need to add another parameter which is tweet_fields
response=client.search_recent_tweets(query=query, max_results=30, tweet_fields=['created_at','lang'])

#let us get only the tweets and display the specified fields
for tweet in response.data:
    print(tweet.id)
    print(tweet.lang)
    print(tweet.created_at)
    print("______")
```

**Tweets Fields**

```python
# You can specify additional Tweet fields to retrieve using tweet_fields
response = client.search_recent_tweets(
    "Tweepy", tweet_fields=["created_at", "lang"]
)
tweets = response.data

# You can then access those fields as attributes of the Tweet objects
for tweet in tweets:
    print(tweet.id, tweet.lang)

# Alternatively, you can also access fields as keys, like a dictionary
for tweet in tweets:
    print(tweet["id"], tweet["lang"])

# Thereâ€™s also a data attribute/key that provides the entire data dictionary
for tweet in tweets:
    print(tweet.data)
```

```python

# Get User's Tweets

# This endpoint/method returns Tweets composed by a single user, specified by
# the requested user ID

user_id = 2244994945

response = client.get_users_tweets(user_id)

# By default, only the ID and text fields of each Tweet will be returned
for tweet in response.data:
    print(tweet.id)
    print(tweet.text)

# By default, the 10 most recent Tweets will be returned
# You can retrieve up to 100 Tweets by specifying max_results
response = client.get_users_tweets(user_id, max_results=30)
```

**Expansions**

```python

# You can specify expansions to retrieve additional objects that relate to the
# returned results
response = client.search_recent_tweets(
    "Tweepy", expansions=["attachments.media_keys", "author_id"]
)
tweets = response.data

# You can then access those objects in the includes Response field
includes = response.includes
users = includes["users"]

# The IDs that represent the expanded objects are included directly in the
# returned data objects
for tweet in tweets:
    print(tweet.author_id)

# An efficient way of matching expanded objects to each data object is to
# create a dictionary of each type of expanded object, with IDs as keys
users = {user["id"]: user for user in users}
for tweet in tweets:
    print(tweet.id, users[tweet.author_id].username)
```

```python
# Get User's Liked Tweets

# This endpoint/method allows you to get information about a userâ€™s liked 
# Tweets

user_id = 2244994945

# By default, only the ID and text fields of each Tweet will be returned
# Additional fields can be retrieved using the tweet_fields parameter
response = client.get_liked_tweets(user_id, tweet_fields=["created_at"])

for tweet in response.data:
    print(tweet.id, tweet.created_at)
```

```python
# Get User's Followers

# This endpoint/method returns a list of users who are followers of the
# specified user ID

user_id = 2244994945

# By default, only the ID, name, and username fields of each user will be
# returned
# Additional fields can be retrieved using the user_fields parameter
response = client.get_users_followers(
    user_id, user_fields=["profile_image_url"]
)

for user in response.data:
    print(user.username, user.profile_image_url)

# By default, this endpoint/method returns 100 results
# You can retrieve up to 1000 users by specifying max_results
response = client.get_users_followers(user_id, max_results=1000)
```
