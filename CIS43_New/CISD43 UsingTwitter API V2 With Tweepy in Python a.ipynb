**Understanding Twitter API access levels**
- https://developer.twitter.com/en/docs/twitter-api/getting-started/about-twitter-api
    - Essential access is available for anybody who signs for it and you do not need to wait for a use case review.
    - You sign up and you get certain functionality on the twitter api v2. You will only get access from tweets from the last seven days.
    - You can get up to 500 thousand tweets per month. That basically means that when you retrieve tweets you can get up to 500 thousand tweets per month as part of this essential access.
    - If you want to get more you can apply to use a case review and according to your use case, you can get access to the elevated package. You can get up to two million tweets per month.
    - If you are a research student you can apply for a use case review for academic research and once you are approved, you get up to 10 million tweets per month. You also get access to the full archive search since 2006.

- How to find and regenerate your API Key and Secret after App creation
- If you've already created an App and need to find or regenerate your API Key and Secret, please follow these steps:

- Navigate to the developer portal
- https://developer.twitter.com/en/portal/projects-and-apps
- Expand the 'Projects and Apps' dropdown in the sidenav
- Open the App which is associated with the API Key and Secret that you would like to find or regenerate
- Navigate to the Keys and tokens tab
- From there, you will find all of the credentials associated with your App.
